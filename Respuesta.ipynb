**4. AnÃ¡lisis Exploratorio de Datos**

**4.1. Que porcentaje de pedidos estan autorizados?**

```python
#Crear una columna para identificar pedidos subautorizados
df_fot['subauthorized'] = df_fot['products_total']< df_fot['purchase_total_price']
df_fot
```

```python
#Calcular el porcentaje de pedidos subautorizados
subauthorized_percentage = (df_fot['subauthorized'].sum()/len(df_fot))*100
print(f"Porcentaje de pedidos subautorizados: {subauthorized_percentage:.2f}%")
```

**4.2. QuÃ© porcentajes de pedidos se autorizan correctamente con una autorizaciÃ³n incremental (+20%) sobre el monto en el checkout?**

```python
#Crear una columna para la autorizaciÃ³n del 20%
df_fot['20_increase'] = df_fot['products_total'] * 1.20
#Verificar si la autorizacion incremental cubre el precio de compra
df_fot['is_authorized'] = df_fot['20_increase'] >= df_fot['purchase_total_price']
#Calcular el porcentaje de pedidos correctamente autorizados con una autorizaciÃ³n incremental del 20%.
correctly_authorized_percentage = (df_fot['is_authorized'].sum()/len(df_fot)) * 100
print(f'Porcentaje de pedidos correctamente autorizados con autorizaciÃ³n incremental (20 %): {correctly_authorized_percentage:.2f}%')
```

**4.3. Hay diferencias cuando se dividen por paÃ­s?**

```python
#### Calcular el % de pedidos subautorizados y correctamente autorizados por paÃ­s.
country_stats = df_fot.groupby(' country_code').agg(
    total_orders=('order_id','count'),
    subauthorized_orders=('subauthorized','sum'),
    correctly_authorized_orders=('is_authorized','sum'))
#Calcular los %
country_stats['subauthorized_percentage'] = (country_stats['subauthorized_orders']/country_stats['total_orders'])*100
country_stats['correctly_authorized_percentage'] = (country_stats['correctly_authorized_orders']/country_stats['total_orders'])*100

print(country_stats[['subauthorized_percentage','correctly_authorized_percentage']])
```

ConclusiÃ³n: Si, hay diferencias significativas cuando se dividen los pedidos por paÃ­s. Los % de pedidos subautorizados y correctamente autorizados varÃ­an notablemente segÃºn la regiÃ³n, lo cual sugiere una fluctuaciÃ³n en la relaciÃ³n entre el product_total y el purchase_total_price.

**4.4. Para el resto de pedidos que quedaria fuera de la autorizaciÃ³n incremental, Â¿quÃ© valores serÃ­an necessarios para capturar el monto restante?**

**3. Experimento**

**3.1. Tipo de test**

Realizariamos una test A/B donde dividiria a los usuarios en dos grupos aleatorios:
- Grupo de Control: Tarifa de entrega actual de 1.9â‚¬
- Grupo de Experimento: Nueva tarifa de entrega de 2.1â‚¬

**3.2. PÃºblico del test**

Aplicaria el test a todos los usuarios activos, ya que el impacto de un tarifa de entrega afecta la decisiÃ³n de compra de ambos segmentos y 
podrÃ­a tener diferentes efectos en cada grupo.

**3.3. Suposiciones**

- Precio: Asumimos que el incremento de la tarifa podria reducir la frecuencia de pedidos o el gasto promedio por cliente.
- Comportamiento Nuevos vs Recurrentes: Los usuarios recurrentes pueden ser menos sensibles al cambio, mientras que los nuevos podrian ser mÃ¡s 
propensos a cancelar el pedido.

**3.3. ValidaciÃ³n**

Usar datos historicos para realizar analisis deL precio antes de implementar el experimento, para tener una idea inicial de cÃ³mo
podria responder cada grupo.

**3.4. DeterminaciÃ³n de la duraciÃ³n del experimento**

Para determinar la muestra del experimento realizaremos divisiones aleatorias para prever el EMD en cada nivel. 

**3.5. KPIs para evaluar el exito**

- Pedidos
- Ingreso promedio por pedido
- Valor promedio de producto por pedido
- Margen de tarifa relativa

**3.6. AnÃ¡lisis de resultados**

Realizar pruebas para comparar cada KPI en ambos grupos. Usar t-tests, para ver si existe 
    una diferencia significativa entre los dos grupos.


**3.7. Recomendaciones**
- Si el aumento de la tarifa no afecta significativamente los KPIs clave o tiene un impacto positivo, 
recomendaria implementar el aumento de tarifa de 2.1â‚¬
- Si el aumento reduce significativamente la frecuencia de pedidos, recomendaria:
    - Segmentar la tarifa: Aplicar el aumento solo para ciertos segmentos menos sensibles al precio, como usuarios con alta frecuencia de pedidos.
    - Ofrecer incentivos a nuevos usuarios: Si los usuarios nuevos muestran una mayor sensibilidad al aumento, considerar tarifas promocionales
    podria ayudar a mejorar la tasa de retenciÃ³n.

```python
#Division los usuarios en dos grupos aleatorios y asignar al grupo de control la tarifa 1.9â‚¬ y al grupo experimento 2.1â‚¬
df_fot['grupo'] = np.random.choice(['control','experimento'], size=len(df_fot), p = [0.5, 0.5])
df_fot['delivery_fee'] = np.where(df_fot['grupo'] == 'control', 1.9, 2.1)
```

```python
#Definir KPIs

#Ingreso promedio por pedido
average_revenue_control = df_fot[df_fot['grupo'] == 'control']['purchase_total_price'].mean()
average_revenue_experimento = df_fot[df_fot['grupo'] == 'experimento']['purchase_total_price'].mean()

#Margen de tarifa relativa
df_fot['relative_margin'] = ((df_fot['purchase_total_price'] - df_fot['products_total']) / df_fot['products_total']) * 100
relative_margin_control = df_fot[df_fot['grupo'] == 'control']['relative_margin'].mean()
relative_margin_experimento = df_fot[df_fot['grupo'] == 'experimento']['relative_margin'].mean()

#Valor promedio de producto por pedido
df_fot['average_product_value'] = df_fot['purchase_total_price']/df_fot['products']
average_product_value_control = df_fot[df_fot['grupo'] == 'control']['average_product_value'].mean()
average_product_value_experimento = df_fot[df_fot['grupo'] == 'experimento']['average_product_value'].mean()
```

```python
#Pruebas estadisiticas entre grupos
from scipy.stats import ttest_ind, ks_2samp
#T-test para comparar ingresos promedio
t_stat_revenue, p_value_revenue = ttest_ind(
    df_fot[df_fot['grupo'] == 'control']['purchase_total_price'],
    df_fot[df_fot['grupo'] == 'experimento']['purchase_total_price'],
    equal_var=False)

#T-test para comparar el Margen de tarifa relativa
t_stat_margin, p_value_margin = ttest_ind(
    df_fot[df_fot['grupo'] == 'control']['relative_margin'],
    df_fot[df_fot['grupo'] == 'experimento']['relative_margin'],
    equal_var= False)

#T-test para comparar el Valor promedio de producto por pedido
t_stat_product_value, p_value_product_value = ttest_ind(
    df_fot[df_fot['grupo'] == 'control']['average_product_value'],
    df_fot[df_fot['grupo'] == 'experimento']['average_product_value'],
    equal_var=False)
```

```python
#Resultados

results = {
    'Ingreso Promedio por Pedido': {'control_mean': average_revenue_control, 'experimento_mean': average_revenue_experimento, 'p_value': p_value_revenue},
    'Margen de Tarifa Relativa': {'control_mean': relative_margin_control, 'experimento_mean': relative_margin_experimento, 'p_value': p_value_margin},
    'Valor Promedio de Productos por Pedido': {'control_mean': average_product_value_control, 'experimento_mean': average_product_value_experimento, 'p_value': p_value_product_value},
}

# Filtrar las mÃ©tricas que resultaron significativas (p-value < 0.05)
significant_results = {metric: stats for metric, stats in results.items() if stats['p_value'] < 0.05}
results, significant_results
```
